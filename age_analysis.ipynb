```python
title_list = [' vifsa_name',' p0_4',' p5_9',' p10_14',' p15_19',' p20_24', ' p25_29', ' p30_34', ' p35_39',' p40_44', ' p45_49',' p50_54', ' p55_59',' p60_64', ' p65_69', ' p70_74', ' p75_79', 'p80_84', ' p85_ov']
new_data = age_data[title_list]

```

```python
new_data.head(5)
```

```python
youth_data = new_data[[' p15_19',' p20_24', ' p25_29']]
                              
                              
                              
```

```python
youth_data['total'] = youth_data.sum(axis=1)
```

```python
youth_data['total']
```

```python
old_data = new_data[[' p65_69', ' p70_74', ' p75_79', 'p80_84', ' p85_ov']]
```

```python
old_data['total'] = old_data.sum(axis=1)
```

```python
dis_dict = {}

for i in range(len(youth_data)):
    dis = new_data[' vifsa_name'][i]
    total_num = youth_data['total'][i] + old_data['total'][i]
    youth_pro = youth_data['total'][i] / total_num
    dis_dict[dis] = youth_pro
```
