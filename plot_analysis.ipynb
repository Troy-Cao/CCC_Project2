```python
import pandas as pd
import json
import matplotlib.pyplot as plt

with open('dis_age.json', 'r') as file:
    age_data = json.load(file)
```

```python
name_list = age_data.keys()
value_list = age_data.values()
```

```python
plt.bar(name_list, value_list, width = 0.35, color='0.4')
plt.ylim(0.7, 1)
plt.xticks(rotation=270)
for a,b in zip(name_list,value_list):  
     plt.text(a, b+0.01, '%.3f' % b, ha='center', va= 'bottom',fontsize=8)  
plt.xlabel("Melbourne District",fontsize=13,fontweight='bold')
plt.ylabel("Youth Proportion",fontsize=13,fontweight='bold')
```
